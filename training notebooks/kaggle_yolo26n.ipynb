```python
# 1. Install Libraries
!pip install -q "ultralytics>=8.4.0" roboflow

import os
from roboflow import Roboflow
from ultralytics import YOLO

# 3. Path Setup
# Change 'my-old-weights.pt' to whatever your uploaded file is called
weights_path = '/kaggle/input/datasets/<DATASET_NAME>/best-nano-pt/best_nano.pt'
```

```python
!pip install roboflow

from roboflow import Roboflow
rf = Roboflow(api_key="YOUR_API_KEY")
project = rf.workspace("madhavs-workspace-507zc").project("indiangrocerymgmtcv")
version = project.version(4)
dataset = version.download("yolov8")
```

```python
# 4. Start Training
model = YOLO(weights_path)

results = model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    optimizer="MuSGD",
    lr0=0.001,  # Low LR to maintain the 94% foundation
    save_period=5,
    project="/kaggle/working/",
    name="yolo26n_full_run",
    cache=True,  # Speeds up training on Kaggle
)
```

```python
import shutil
import os
from IPython.display import FileLink

# 1. Define the folder you want to zip (the 'name' you used in model.train)
folder_to_zip = "/kaggle/working/yolo26n_full_run"
output_zip = "/kaggle/working/final_training_results"

# 2. Create the ZIP file
shutil.make_archive(output_zip, "zip", folder_to_zip)

# 3. Create a clickable download link in the console
print("Training Complete! Click the link below to download your results:")
FileLink(r"final_training_results.zip")
```
